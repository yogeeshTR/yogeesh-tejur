```python
'''2. Write a program which will find all such numbers which are divisible by 7 but are not a multipleof 5, 
between 2000 and 3200 (both included). The numbers obtained should be printed in acomma-separated sequence on a single line.'''

#Accepting first number from the user
first=int(input("enter the first number :" ))

#Accepting the last value from user.
last=int(input("enter the last number :" ))

#lasr value gives n-1 value hence here we uppend by n+1
a=list(range(first,last+1))


c=[]                    #create an empty list
for i in a:             # here for loop used to itrate each numbers in given range of values.
    if i%7==0:           
        if i%5==0:       # To the multiples of 5 from the given range off values.
            continue     #it skips the multiples  of 5 in the range
        c.append(i)
print(c)    
        
```

```python
#2.Create the below pattern using nested for loop in Python.

n=int(input())     # here we take the max number of star from user

#itrate each value from zero to n+1 value.
for i in range(0,n+1): 
    print("*"*i)                            # multiply each number with sigh "*".
    if i==n:                               # if i==n then start value in reverse scaling.
        for j in range(n+1,0,-1):          # applying reverse scaling method to 
            print("*"*j)
```
